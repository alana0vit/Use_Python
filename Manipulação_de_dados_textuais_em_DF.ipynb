```python
# URL: podemos também acessar os dados direto pela URL sem a necessidade de fazer download

import pandas as pd
dados1 = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv")

dados1.head()
```

```python
# 2. Variáveis
# Para simplificar a visualização, irei usar apenas as colunas ‘Name’ e ‘Sex’:
# Para selecionar essas colunas, vou usar o parâmetro usecols ao ler os dados, com os nomes das colunas em uma lista:

import pandas as pd
dados2 = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv',
                      usecols = ['Name', 'Sex'])

dados2.head()
```

```python
# 3. Str accessor
# Para manipularmos os valores textuais, iremos usar o str accessor, que torna os métodos de strings acessíveis para serem utilizados nos valores das colunas do nosso DataFrame,
# já que cada coluna isolada é interpretada como uma Series:

type(dados1['Name'])

# Basicamente, usaremos a seguinte construção para aplicar os métodos de strings: dados[‘Coluna’].str.metodo()
# a) dados[‘Coluna’]: selecionar a coluna de interesse
# b) .str: adicionar o str accessor
# c) .metodo(): aplicar o método
```

```python
# 4. Exemplos
# Para deixar os caracteres da coluna ‘Name’ em minúsculo, usamos o lower():

dados1['Name'].str.lower()
```

```python
# Vamos criar uma nova coluna chamada ‘Name_lowercase’ para capturar esses nomes em minúsculo:

dados1['Name_lowercase'] = dados1["Name"].str.lower()
dados1.head()
```

```python
# 4.2 Split()
# Na coluna ‘Name’, o sobrenome está separado do nome pela vírgula. Vamos usar o split() para criar uma lista com dois elementos:
# [0] Sobrenome do passageiro, [1] Nome do passageiro
# Usando como separador a vírgula:

dados1['Name'].str.split(',')
```

```python
# split(expand = True)
# Você pode passar True para o parâmetro expand do split() para retornar um DataFrame com os valores que foram separados, cada um indo para uma coluna:

dados1['Name'].str.split(',', expand = True)
```

```python
# 4.3 Get()
# Aqui, vamos extrair o sobrenome (índice zero) da lista criada anteriormente quando usamos split(), encadeando o método get():

dados1['Name'].str.split(',').str.get(0)
```

```python
# Vou criar uma nova coluna ‘Surname’ com os sobrenomes extraídos:

dados1['Surname'] = dados1['Name'].str.split(',').str.get(0)
dados.head()
```

```python
# 4.4 Contains()
# Esse método checa se cada valor da coluna selecionada contém ou não uma dada string, retornando True ou False. Podemos usar essa saída booleana para criar um subset dos nossos
# dados.
# Vou buscar quais linhas contém na coluna ‘Name_lowercase’ o nome ‘elizabeth’:

dados1[dados1['Name_lowercase'].str.contains('elisabeth')]
```

```python
# 4.5 Startswith()
# Podemos checar quais valores começam com uma dada string usando o startswith(), que retorna True ou False.
# Vamos ver quais valores tem sobrenome que começam com ‘Go’:

dados1[dados1['Surname'].str.startswith('Go')]
```

```python
# Mostrando a diferença de adicionar ou não o nome do dataframe
dados1['Surname'].str.startswith('Go')
```

```python
# 4.6 Len()
# Também podemos obter o tamanho de cada elemento, aqui no caso, o tamanho dos sobrenomes:

dados1['Surname'].str.len()
```

```python
# E onde está o sobrenome que tem o maior tamanho? Podemos usar o idxmax() para retornar o índice da linha que possui o maior valor numérico.
# O idxmax() não é um método de strings, já que ele checa pelo maior valor numérico, por isso não precisamos adicionar o str antes dele:

maior_indice = dados1['Surname'].str.len().idxmax()
print(maior_indice)

maior_indice_str= dados1.loc[430, 'Surname']
print(maior_indice_str)

# Logo, o maior sobrenome é ‘Bjornstrom-Steffansson’
```

```python
# 4.7 Replace()
# Também podemos substituir um valor por outro usando o replace().
# O método replace() também não é um método de strings, por isso não precisamos adicionar o str obrigatoriamente antes.
# Vamos substituir na coluna ‘Sex’, os valores ‘male’ por ‘M’, e ‘female’ por ‘F’, usando um dicionário:

dados1['Sex'].replace({'female': 'F', 'male': 'M'})
```

```python
# Podemos criar uma nova coluna para abrigar esses novos valores:

dados1['Sex_short'] = dados1['Sex'].replace({'female': 'F', 'male': 'M'})
dados1.head()
```
